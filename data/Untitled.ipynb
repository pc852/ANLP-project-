```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
import sklearn as sk
import os
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from numpy.random import uniform
from sklearn.feature_extraction.text import CountVectorizer
import unidecode
import pandas as pd
import nltk
import re
from os import listdir
from os.path import isfile, join
```

```python
"""read file"""
def read_file(file_name):
    file = unidecode.unidecode(open(file_name).read())
    return file

def read_csv(file_name, sep=','):
    df = pd.read_csv(file_name, sep=sep)
    return df

def generate_BoW(corpus):
    vectorizer = CountVectorizer()
    corpus_vectorized = vectorizer.fit(corpus)
    return corpus_vectorized
```

```python
corpus = []
initials = ['19','a','b','c','d','e','m1']#,'f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
file_root = 'Dataset/AZlyrics/azlyrics_lyrics_'
file_format = '.csv'
for initial in initials:
    file_name = file_root + initial + file_format
    file = read_csv(file_name)
    
    for sent in file['LYRICS'].values:
        sent = sent.lower()
        sent = re.sub(r'[^a-zA-Z0-9_\']',' ',sent)
        sent = re.sub(r'[\s,.?]+',' ',sent)
        word_list = nltk.tokenize.SpaceTokenizer().tokenize(sent)
        for word in word_list:
            corpus.append(word)
corpus = set(corpus)
```

```python
print(len(corpus))
corpus_vectorized = generate_BoW(corpus)
print (corpus_vectorized)
print(corpus_vectorized.vocabulary_.get('king'))
```

```python
train_rap_dir = 'Dataset/Training/Rap/'
train_nrap_dir = 'Dataset/Training/Not Rap/'
rap_training_set = [f for f in listdir(train_rap_dir) if isfile(join(train_rap_dir, f))]
nrap_training_set = [f for f in listdir(train_nrap_dir) if isfile(join(train_nrap_dir, f))]

train_rap_list = []
train_nrap_list = []


for file in rap_training_set:
    data = unidecode.unidecode(open(join(train_rap_dir,file)).read())
    data = data.lower()
    data = re.sub(r'[^a-zA-Z0-9_\']',' ',data)
    data = re.sub(r'[\s,.?]+',' ',data)
    train_rap_list.append(data)

rap_training_set = []
rap_training_set = corpus_vectorized.transform(train_rap_list)


for file in nrap_training_set:
    data = unidecode.unidecode(open(join(train_nrap_dir,file)).read())
    data = data.lower()
    data = re.sub(r'[^a-zA-Z0-9_\']',' ',data)
    data = re.sub(r'[\s,.?]+',' ',data)
    train_nrap_list.append(data)

nrap_training_set = []
nrap_training_set = corpus_vectorized.transform(train_nrap_list)


X_train = []
Y_train = []

for i in range(rap_training_set.shape[0]):
    X_train.append(rap_training_set.toarray()[i])
    Y_train.append(1)
for i in range(nrap_training_set.shape[0]):
    X_train.append(nrap_training_set.toarray()[i])
    Y_train.append(0)
print(X_train)
print(len(X_train))
print(Y_train)
print(len(Y_train))
```
